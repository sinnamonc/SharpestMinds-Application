```python
import numpy as np
import pandas as pd
import time
import math
import sys
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Dropout
from keras.layers.normalization import BatchNormalization
from keras.optimizers import RMSprop
import keras.initializers
import keras.callbacks
import matplotlib.pyplot as plt
from data_preprocessing import import_data
from utils import shuffle_in_unison
from utils import percent_correct
from utils import get_uniform_batch
from data_preprocessing import get_data_sets
print('Complete')
```

A comparison of some neural nets to see if there is an appreciable different if the data columns have mean zero or not.

```python
test_data_z, test_labels_z, valid_data_z, valid_labels_z, train_data_z, train_labels_z = get_data_sets(mean_zero = True)
test_data, test_labels, valid_data, valid_labels, train_data, train_labels = get_data_sets(mean_zero = False)
print('Complete')
```

```python
t_init = time.time()
print('Model in training...')
modl = Sequential()
modl.add(Dense(128, activation='relu', input_dim=54))
modl.add(BatchNormalization())
modl.add(Dense(64, activation='relu'))
modl.add(BatchNormalization())
modl.add(Dense(7, activation='softmax'))

optimizer = keras.optimizers.Adam()
# model.compile(loss='mean_squared_error', optimizer=optimizer)
modl.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=["accuracy"])
# train the model several epochs, and test on the validation set. Plot the loss for train and validation sets

early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=200)

history = modl.fit(train_data, train_labels, batch_size = 100000, epochs = 2500, 
                    validation_data=(valid_data,valid_labels), verbose = 1, callbacks=[early_stopping])

print('Model created in = {:0.1f} minutes.'.format((time.time()-t_init)/60))
      
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

plt.plot(history.history["acc"])
plt.plot(history.history["val_acc"])
plt.title('model acc')
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```
